```python
d = [df['interveningOpponents'][k] + df['interveningTeammates'][k] for k in range(8925)]
new_col = pd.DataFrame({"intervening": d})
df = df.join(new_col['intervening'])
df = df.drop(['interveningOpponents', 'interveningTeammates'], axis = 1)
df
```

```python
d = [df_test['interveningOpponents'][k] + df_test['interveningTeammates'][k] for k in range(2000)]
new_col = pd.DataFrame({"intervening": d})
df_test = df_test.join(new_col['intervening'])
df_test = df_test.drop(['interveningOpponents', 'interveningTeammates'], axis = 1)
df_test
```

```python
X_train = df[['playType', 'bodyPart', 'interferenceOnShooter','x','y' ,'intervening', 'minute']]
Y_train = df[['outcome']]
X_test = df_test[['playType', 'bodyPart', 'interferenceOnShooter','x','y' ,'intervening', 'minute']]
Y_train
```

```python
X_train = pd.get_dummies(X_train, columns = ['playType','bodyPart','interferenceOnShooter'])
X_test = pd.get_dummies(X_test, columns = ['playType','bodyPart','interferenceOnShooter'])
```

```python
clf = LogisticRegression(random_state=0, C = 0.09, class_weight = 'balanced').fit(X_train, Y_train)
```

```python
predictions = clf.predict_proba(X_test)
```

```python
predictions
```

```python
result = pd.DataFrame(predictions, columns = ['1','prediction'])
result = result[['prediction']]
result
```
